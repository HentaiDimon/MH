```python
#1. Створіть одновимірний масив (вектор) з оберненими значеннями факторіалів натуральних чисел, починаючи з 2, та виведіть його значення.
import math

# Створити список чисел від 2 до 10 (включно)
number = list(range(2, 11))

# Обчислити факторіал для кожного числа
factorial = [math.factorial(n) for n in number]

# Обчислити обернене значення для кожного факторіала
inverse_factorial = [1/f for f in factorial]

# Вивести список обернених значень факторіалів
print(inverse_factorial)
```

```python
#2. Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його оберненимизначеннями синусів кожного натурального числа, починаючи з 1, та виведіть йогозначення.
import numpy as np

# Створити список чисел від 1 до 9 (включно) для заповнення матриці 3x3
numbers = list(range(1, 10))

# Обчислити синус для кожного числа та взяти обернене значення
inverse_sines = [1 / np.sin(n) for n in numbers]

# Перетворити список в матрицю 3x3
matrix = np.array(inverse_sines).reshape(3, 3)

# Вивести матрицю
print("Матриця обернених значень синусів:\n", matrix)
```

```python
#Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть її обернену матрицю.
import numpy as np

# Створити матрицю 2x2 з випадковими цілими числами від 1 до 10
matrix = np.random.randint(1, 11, size=(2, 2))

# Знайти обернену матрицю
inverse_matrix = np.linalg.inv(matrix)

print("Оригінальна матриця:\n", matrix)
print("\nОбернена матриця:\n", inverse_matrix)
```

```python
#Створіть дві матриці розміром 3x3, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та знайдіть відношення всіх
#елементів першої матриці до відповідних елементів другої матриці (поелементне ділення).
import numpy as np

# Створити дві матриці 3x3 з випадковими цілими числами від 1 до 5
matrix1 = np.random.randint(1, 6, size=(3, 3))
matrix2 = np.random.randint(1, 6, size=(3, 3))

# Знайти відношення всіх елементів першої матриці до відповідних елементів другої матриці
result = matrix1 / matrix2

print("Матриця 1:\n", matrix1)
print("\nМатриця 2:\n", matrix2)
print("\nРезультат поелементного ділення:\n", result)
```

```python
#Створіть дві матриці розміром 4x4, заповніть їх випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть різницю їх елементів.
import numpy as np

# Створити дві матриці 4x4 з випадковими дійсними числами від 0 до 1
matrix1 = np.random.rand(4, 4)
matrix2 = np.random.rand(4, 4)

# Знайти різницю елементів матриць
difference = matrix1 - matrix2

print("Матриця 1:\n", matrix1)
print("\nМатриця 2:\n", matrix2)
print("\nРізниця матриць:\n", difference)
```

```python
#Створіть матрицю розміром 4x4, заповніть її випадковими дійсними числами в діапазоні від -10 до 10 та знайдіть суму елементів на побічній діагоналі.
import numpy as np

# Створити матрицю 4x4 з випадковими дійсними числами від -10 до 10
matrix = np.random.uniform(-10, 10, size=(4, 4))

# Знайти суму елементів на побічній діагоналі
secondary_diagonal_sum = np.trace(np.fliplr(matrix))

print("Згенерована матриця:\n", matrix)
print("\nСума елементів на побічній діагоналі:", secondary_diagonal_sum)
```

```python
#Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть
#вектор-стовпчик, що містить суму квадратів елементів кожного стовпця матриці
import numpy as np

# Створити матрицю 3x3 з випадковими дійсними числами від 0 до 1
matrix = np.random.rand(3, 3)

# Обчислити суму квадратів елементів кожного стовпця
column_sums_of_squares = np.sum(matrix ** 2, axis=0)

# Перетворити результат на вектор-стовпчик
column_sums_of_squares = column_sums_of_squares.reshape(-1, 1)

print("Матриця:\n", matrix)
print("\nВектор-стовпчик суми квадратів елементів кожного стовпця:\n", column_sums_of_squares)
```
